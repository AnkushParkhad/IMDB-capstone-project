### Now lets scrape rest of the columns using beautiful soup

```python
pages=np.arange(1,1753,50)
pages
```

```python
#creating empty list to store metascore
metascore=[]

for page in pages:
    page=requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup=BeautifulSoup(page.text,'html.parser')
    
    movie_data=soup.findAll('div',class_='lister-item mode-advanced')
    time.sleep(5)
    
    for store in movie_data:
        meta=store.find('span',class_="metascore").text if store.find('span',class_="metascore") else "N/A"
        metascore.append(meta)
```

```python
#checking weather all items appended in list or not
print(len(metascore))
```

```python
#creating empty list to store certification
certification=[]

for page in pages:
    page=requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup=BeautifulSoup(page.text,'html.parser')
    
    movie_data=soup.findAll('div',class_='lister-item mode-advanced')
    time.sleep(5)
    
    for store in movie_data:
        cert=store.find('span',class_="certificate").text if store.find('span',class_="certificate") else "N/A"
        certification.append(cert)
```

```python
#1)  Display all the details of movies created by directors Christopher and Matt Reeves.

results=cur.execute('select Movie_name from IMDB_table_first where Director_1="Christopher Nolan" or "Matt Reeves" or Director_2="Christopher Nolan" or "Matt Reeves" or Director_3="Christopher Nolan" or "Matt Reeves" or Director_4="Christopher Nolan" or "Matt Reeves" or Director_5="Christopher Nolan" or "Matt Reeves" or Director_6="Christopher Nolan" or "Matt Reeves"')
results.fetchall()
```

```python
#2) Display all the details of movies with a duration of 140 minutes to 190 minutes.
results=cur.execute('select * from IMDB_table_first where Duration between "140 min" and "190 min"')
for i in results:
    print(i)
```

```python
#3) Display all details of movies with ratings above 7 in ascending order.

results=cur.execute('select * from IMDB_table_first where Ratings>7 order by Ratings asc')
for i in results:
    print(i)
```

```python
#4) Display all movie names in descending order.

results=cur.execute("select Movie_name from IMDB_table_first order by Serial_no desc")
for i in results:
    print(i)
```

```python
#5) Display movie name starts with ‘P’ and their rating is greater than 7.

results=cur.execute('select Movie_name from IMDB_table_first where Movie_name like "P%" and Ratings>7')
for i in results:
    print(i)
```
